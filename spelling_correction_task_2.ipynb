```python
import pandas as pd
from textblob import TextBlob
import re
import string
```

```python
df = pd.read_csv("/content/spell.csv", header=None)
df
```

```python
df = df.drop_duplicates()
df.shape
```

```python
def clean(text):
    text = str(text).lower()
    text = re.sub('\[.*?\]', '', text)
    text = re.sub('https?://\S+|www\.\S+', '', text)
    text = re.sub('<.*?>+', '', text)
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text)
    text = re.sub('\n', '', text)
    text = re.sub('\w*\d\w*', '', text)
  
    return text

df["processed"] = df[0].apply(clean)
```

```python
df
```

```python
def correction(x):
  ans = TextBlob(x)
  return "".join(ans.correct())
```

```python
df['corrected'] = df['processed'].apply(correction)
```
