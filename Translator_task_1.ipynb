```python
# importing packages:

import pandas as pd
import re
import nltk
import nltk

nltk.download('stopwords')
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('stopwords')
nltk.download('averaged_perceptron_tagger')
```

```python
data = pd.read_csv("/content/English.csv")   # reading the file
data
```

# Data Cleaning:

```python
from nltk.corpus import stopwords
import string
stopword=set(stopwords.words('english'))
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()

def clean(text):
    text = str(text).lower()
    text = re.sub('\[.*?\]', '', text)
    text = re.sub('https?://\S+|www\.\S+', '', text)
    text = re.sub('<.*?>+', '', text)
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text)
    text = re.sub('\n', '', text)
    text = re.sub('\w*\d\w*', '', text)
    text = [word for word in text.split(' ') if word not in stopword]
    text=" ".join(text)
    text = [lemmatizer.lemmatize(word) for word in text.split(' ')]
    text=" ".join(text)
    return text

data["English_processed"] = data["English words/sentences"].apply(clean)
```

```python
data
```

```python
data["English_processed"] = data["English_processed"].drop_duplicates()  # dropping duplicates
data.shape
```

```python
data = data.dropna()   # dropping null values
data.shape
```

```python
df = data["English_processed"].tail(25)     # taking only last 25 words
df
```

```python
df = pd.DataFrame(df)
df = df.reset_index(drop=True)
```

```python
from deep_translator import GoogleTranslator

# function for Translation:

def translate_french(text):
  return GoogleTranslator('auto','fr').translate(text)

def translate_german(text):
  return GoogleTranslator('auto','de').translate(text)

def translate_spanish(text):
  return GoogleTranslator('auto','es').translate(text)
```

```python
df["French"] = df["English_processed"].apply(translate_french)
df["German"] = df["English_processed"].apply(translate_german)
df["Spanish"] = df["English_processed"].apply(translate_spanish)
```
